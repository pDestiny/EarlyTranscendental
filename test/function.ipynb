```python
import matplotlib.pyplot as plt
import numpy as np
```

### 5-(a)
$$y=x^3$$

```python
f = lambda x: (x-2)**3 + 3
```

```python
x = np.linspace(-5, 10, 100)
```

```python
plt.plot(x, f(x))
plt.grid(True)

plt.axhline(y=0, color="black", linestyle="--")
plt.axvline(x=0, color="black", linestyle="--")
plt.show()
```
